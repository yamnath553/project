```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
gdp_data = pd.read_csv('C:/Users/dahal/Downloads/gdp_csv.csv')
```

```python
gdp_data.head(10)
```

```python
print("Initial Dataset Info:")
print(gdp_data.info())
print("\nFirst Few Rows of Dataset:")
print(gdp_data.head())
```

```python
missing_values = gdp_data.isnull().sum()
print("\nMissing Values in Each Column:")
print(missing_values)
```

```python
gdp_data = gdp_data[gdp_data['Value'] > 0]
```

```python
gdp_data
```

```python
print("\nSummary Statistics:")
print(gdp_data.describe())
```

```python
plt.figure(figsize=(10, 6))
country_data = gdp_data[gdp_data['Country Name'] == 'India']
plt.plot(country_data['Year'], country_data['Value'], marker='o')
plt.title('GDP Trend for India')
plt.xlabel('Year')
plt.ylabel('GDP (in USD)')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
country_data = gdp_data[gdp_data['Country Name'] == 'China']
plt.plot(country_data['Year'], country_data['Value'], marker='o')
plt.title('GDP Trend for China')
plt.xlabel('Year')
plt.ylabel('GDP (in USD)')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
country_data = gdp_data[gdp_data['Country Name'] == 'Nepal']
plt.plot(country_data['Year'], country_data['Value'], marker='o')
plt.title('GDP Trend for Nepal')
plt.xlabel('Year')
plt.ylabel('GDP (in USD)')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(gdp_data['Value'], bins=50, kde=True, color='blue')
plt.title('Distribution of GDP Values')
plt.xlabel('GDP Value')
plt.ylabel('Frequency')
plt.show()
```

```python
fig = px.scatter(gdp_data, x='Year', y='Value', color='Country Name',
                 title='GDP Trends for All Countries',
                 labels={'Value': 'GDP Value', 'Year': 'Year'})
fig.show()
```

```python
cleaned_file_path = 'C:/Users/dahal/Downloads/cleaned_gdp_data.csv'
gdp_data.to_csv(cleaned_file_path, index=False)
print(f"Cleaned data saved to {cleaned_file_path}")
```
